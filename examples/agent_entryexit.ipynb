# Cymetric Example for Build and Decommissioning of Prototypes

This example walked through how to plot the number of reactor prototypes built during the simulation. 
The exercise starts by running the given input file, 'recycle.xml' through Cyclus to give an output database 'recycle.sqlite'. 



### Learning Objectives:
* Open a SQLite Database with Cymetric
* Store a metric as a pandas DataFrame
* Manipulate the contents and column names of a DataFrame
* Plot the data in a DataFrame in a bar chart

```python
import cymetric as cym
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Remove any current database with the same name, run cyclus to have the most updated output database
! rm recycle.sqlite
! cyclus recycle.xml -o recycle.sqlite #-o flag specifies output file name
```

```python
db = cym.dbopen('recycle.sqlite')
evaler = cym.Evaluator(db)
comm = evaler.eval('BuildSeries')
decomm = evaler.eval('DecommissionSeries') 
```

A full list of metrics avilable through Cymetric can be found in [cymetric/metrics.py](https://github.com/cyclus/cymetric/blob/master/cymetric/metrics.py) in the cymetric repository. General metrics being on line 94 of the code. 

```python
#make the exit counts negative for plotting purposes
neg = -decomm['Count']
decomm = decomm.drop('Count',axis=1)
decomm = pd.concat([decomm, neg],axis=1)
```

```python
#Renaming time columns to facilitate later merging
comm = comm.rename(columns={'EnterTime':'Time'})
decomm = decomm.rename(columns={'ExitTime':'Time'})
```

```python
#Pivot tables to facilitate plotting, then merge tables
c = comm.pivot('Time','Prototype')['Count'].reset_index()
d = decomm.pivot('Time','Prototype')['Count'].reset_index()
cd = pd.merge(c,d,left_on='Time',right_on='Time',how='outer',sort=True, \
              suffixes=('_enter','_exit')).fillna(0)
```

```python
#pandas always changes everything to floats in the presence of NaN, so changing TimeStep values to integers
cd.Time = cd.Time.astype(int)
```

```python
#Plotting facilities of interst: lwr and frx
colors = ['green','purple']
ax = cd[['Time','lwr_enter','frx_enter']].plot.bar(x='Time',figsize=(20,10),stacked=True, \
                                               legend=False, color=colors)
l = plt.legend()
l.get_texts()[0].set_text('LWR')
l.get_texts()[1].set_text('FR')
l.set_title('Prototype')
cd[['Time', 'lwr_exit','frx_exit']].plot.bar(x='Time', ax=ax, figsize=(20,10),stacked=True, \
                                               legend=False, color=colors)
plt.xlabel(r'TimeStep',fontsize=18); plt.ylabel(r'Count',fontsize=18)
plt.show()
#plt.savefig('ex.png')
```
