# Cymetric Example for Build and Decommissioning of Prototypes

This example walked through how to plot the number of reactor prototypes built during the simulation. 
The exercise starts by running the given input file, 'recycle.xml' through Cyclus to give an output database 'recycle.sqlite'. 



```python
import cymetric as cym
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
db = cym.dbopen('recycle.sqlite')
evaler = cym.Evaluator(db)
```

```python
# Store the tables as pandas DataFrames
comm = evaler.eval('BuildSchedule')
decomm = evaler.eval('DecomSchedule') 
```

```python
#Renaming time columns for later merging
comm = comm.rename(columns={'BuildTime':'Time'})
decomm = decomm.rename(columns={'DecomTime':'Time'})
```

```python
c = comm.groupby('Time').size().sort_index().reset_index(name='Prototypes')
```

```python
c.plot.bar(x='Time',legend=False)
```
